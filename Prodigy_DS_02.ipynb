```python
from sklearn.ensemble import BaggingClassifier
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming 'data' is your dataset with features and labels
# Splitting data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data.drop(columns=['Ticket']), data['Embarked'], test_size=0.2, random_state=42)

# Define preprocessing steps for numeric features
numeric_features = data.select_dtypes(include=[np.number]).columns.tolist()
numeric_transformer = make_pipeline(SimpleImputer(strategy='mean'), StandardScaler())

# Combine preprocessing steps for all features using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ])

# Define your model2
model2 = BaggingClassifier(base_estimator=SVC(C=3))

# Initialize pipeline with preprocessing and model
pipe5 = make_pipeline(preprocessor, model2)

# Now you can fit your pipeline and make predictions
pipe5.fit(x_train, y_train)
pred5 = pipe5.predict(x_test)
```
